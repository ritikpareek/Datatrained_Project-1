```python
from scipy.stats import zscore
```

```python
z=np.abs(zscore(df))
```

```python
df=df[(z<3).all(axis=1)]
```

```python
df.shape
```

```python
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
y.shape
```

# linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
max_score=0
for r in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=r)
    lm=LinearRegression()
    lm.fit(x_train,y_train)
    pred=lm.predict(x_test)
    rs=r2_score(y_test,pred)
    print('r_2 score of at',r,'is',rs)
    if rs>max_score:
        max_score=rs
        random_state=r
        
print('\n')
print(max_score)
print(random_state)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=82)
lm.fit(x_train,y_train)
pred=lm.predict(x_test)
```

```python
r2_score(y_test,pred)
```

```python
mean_absolute_error(y_test,pred)
```

```python
print("rmse",np.sqrt(mean_squared_error(y_test,pred)))
```

```python
print(lm.coef_)
print('\n')
print(lm.intercept_)
```

# ElasticNet
